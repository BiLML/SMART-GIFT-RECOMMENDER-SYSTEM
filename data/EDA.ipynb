```python
import re # Biểu thức chính quy
import unicodedata 
import pandas as pd

df = pd.read_json('du_lieu_sach_fahasa.json')
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['price'] = df['price'].fillna(round(df['price'].mean(), 0))

LOWER = 'a-zàáâãèéêìíòóôõùúýăđĩũơưạảấầẩẫậắằẳẵặẹẻẽếềểễệỉịọỏốồổỗộớờởỡợụủứừửữựỳỵỷỹ'
UPPER = 'A-ZÀÁÂÃÈÉÊÌÍÒÓÔÕÙÚÝĂĐĨŨƠƯẠẢẤẦẨẪẬẮẰẲẴẶẸẺẼẾỀỂỄỆỈỊỌỎỐỒỔỖỘỚỜỞỠỢỤỦỨỪỬỮỰỲỴỶỸ'

def clean(text):
    if type(text) != str:
        return ""

    text = unicodedata.normalize('NFC', text)

    # Thay thế ký tự 
    text = text.replace('\n', ' ').replace('\r', ' ').replace('\t', ' ')
    # Tách các từ dính liền sau dấu câu
    text = re.sub(r'(?<=[.,!?])(?=[^\s])', r' ', text)
    # Tách chữ thường dính chữ hoa    
    text = re.sub(fr'([{LOWER}])([{UPPER}])', r'\1 \2', text)
    # Xóa khoảng trắng thừa
    text = re.sub(r'\s+', ' ', text)
    # Xóa khoảng trắng ở đầu và cuối câu
    return text.strip()

df['description'] = df['description'].apply(clean)
df.to_csv('du_lieu_sach_fahasa_clean.csv', index=False, encoding='utf-8-sig')
```
